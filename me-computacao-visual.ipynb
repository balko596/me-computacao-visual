```python
import cv2
import numpy as np
from collections import namedtuple
import matplotlib.pyplot as plt
```

## Definições

```python
WINDOW_WIDTH = 640
WINDOW_HEIGHT = 640
WINDOW_SIZE = (WINDOW_HEIGHT, WINDOW_HEIGHT, 3)
Object = namedtuple("Object", "x y")
```

```python
# define the five objects
obj_1 = Object(320, 320)
obj_2 = Object(0, 0)
obj_3 = Object(0, 0)
obj_4 = Object(0, 0)
obj_5 = Object(0, 0)
```

```python
# define colors
RED = (0, 0, 255)
```

```python
canvas = np.ones(WINDOW_SIZE, dtype="uint8") * 40
```

## Funções com os desenhos

```python
def draw_obj_1(x:int, y:int, scale:int):
  center = (x, y - 8*scale)
  r = 6*scale
  cv2.circle(canvas, center, r, RED, cv2.FILLED)
```
