## Funções com os desenhos

```python
def draw_obj_1(x:int, y:int, scale:int):
  pos_1 = (x - 4*scale, y - 8*scale)
  pos_2 = (x + 4*scale, y + 8*scale)
  center = (x, y - 8*scale)
  r = 6*scale
  cv2.circle(canvas, center, r, RED, cv2.FILLED)
  cv2.rectangle(canvas, pos_1, pos_2, RED, cv2.FILLED)
```
